```python
import os
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, confusion_matrix
from sklearn.manifold import TSNE
from IPython.display import display
from thop import profile

def set_seed(seed: int = 42):
    import random
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = False
    torch.backends.cudnn.benchmark = True
set_seed(42)

def read_txt_matrix(file_path):
    try:
        return np.loadtxt(file_path, dtype=np.float32)
    except Exception as e:
        print(f"Error reading {file_path}: {e}")
        raise

def load_split(root, split):
    UCI_CHANNELS_PREFIX = [
        "total_acc_x_", "total_acc_y_", "total_acc_z_",
        "body_acc_x_", "body_acc_y_", "body_acc_z_",
        "body_gyro_x_", "body_gyro_y_", "body_gyro_z_",
    ]
    channels = []
    for prefix in UCI_CHANNELS_PREFIX:
        file_path = os.path.join(root, f"{prefix}{split}.txt")
        if not os.path.exists(file_path):
            raise FileNotFoundError(f"File not found: {file_path}")
        channels.append(read_txt_matrix(file_path))
    X = np.stack(channels, axis=1)
    y = read_txt_matrix(os.path.join(root, f"y_{split}.txt")).astype(int) - 1
    subjects = read_txt_matrix(os.path.join(root, f"subject_{split}.txt")).astype(int)
    activity_names = ['Walking', 'Walking Upstairs', 'Walking Downstairs', 'Sitting', 'Standing', 'Laying']
    return X, y, subjects, activity_names

class HARDataset(Dataset):
    def __init__(self, X, y, normalize=True):
        self.X = X.astype(np.float32)
        self.y = y.astype(np.int64)
        if normalize:
            self.mean = self.X.mean(axis=(0, 2), keepdims=True)
            self.std = self.X.std(axis=(0, 2), keepdims=True) + 1e-8
            self.X = (self.X - self.mean) / self.std
        else:
            self.mean, self.std = 0, 1

    def set_stats(self, mean, std):
        self.mean, self.std = mean, std
        self.X = (self.X - self.mean) / self.std

    def __len__(self):
        return len(self.y)

    def __getitem__(self, idx):
        return torch.from_numpy(self.X[idx]), int(self.y[idx])

class CompactEncoder(nn.Module):
    def __init__(self):
        super().__init__()
        self.stage1 = self._make_stage(9, 16, stride=2)
        self.stage2 = self._make_stage(16, 32, stride=2)
        self.stage3 = self._make_stage(32, 64, stride=2)

    def _make_stage(self, in_ch, out_ch, stride=1):
        return nn.Sequential(
            nn.Conv1d(in_ch, out_ch, kernel_size=3, stride=stride, padding=1),
            nn.BatchNorm1d(out_ch),
            nn.GELU(),
            nn.Conv1d(out_ch, out_ch, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm1d(out_ch),
            nn.GELU(),
            nn.Conv1d(out_ch, out_ch, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm1d(out_ch),
            nn.GELU()
        )

    def forward(self, x):
        x = self.stage1(x)
        x = self.stage2(x)
        x = self.stage3(x)
        return x

class CompactDecoder(nn.Module):
    def __init__(self):
        super().__init__()
        self.stage1 = self._make_upsample(64, 32)
        self.stage2 = self._make_upsample(32, 16)
        self.stage3 = self._make_upsample(16, 9)

    def _make_upsample(self, in_ch, out_ch):
        return nn.Sequential(
            nn.ConvTranspose1d(in_ch, out_ch, kernel_size=4, stride=2, padding=1),
            nn.BatchNorm1d(out_ch),
            nn.GELU(),
            nn.Conv1d(out_ch, out_ch, kernel_size=3, padding=1),
            nn.BatchNorm1d(out_ch),
            nn.GELU()
        )

    def forward(self, h):
        x = self.stage1(h)
        x = self.stage2(x)
        x = self.stage3(x)
        return x

class CausalManifoldHAR(nn.Module):
    def __init__(self, use_film=True, use_decoder=True):
        super().__init__()
        self.use_film = use_film
        self.use_decoder = use_decoder
        self.encoder = CompactEncoder()
        self.film = nn.Sequential(nn.Linear(6, 32), nn.GELU(), nn.Linear(32, 128)) if use_film else None
        self.to_Z = nn.Conv1d(64, 64, 1)
        self.to_N = nn.Conv1d(64, 64, 1)
        self.classifier = nn.Sequential(
            nn.AdaptiveAvgPool1d(1),
            nn.Flatten(),
            nn.Linear(64, 128),
            nn.GELU(),
            nn.Dropout(0.1),
            nn.Linear(128, 64),
            nn.GELU(),
            nn.Dropout(0.1),
            nn.Linear(64, 6)
        )
        self.decoder = CompactDecoder() if use_decoder else None

    def forward(self, x, psi=None):
        h = self.encoder(x)
        if self.use_film and self.film is not None and psi is not None:
            params = self.film(psi).view(-1, 2, 64, 1)
            gamma, beta = params[:, 0], params[:, 1]
            h = h * gamma + beta
        h_z = self.to_Z(h)
        h_n = self.to_N(h)
        logits = self.classifier(h_z)
        x_recon = self.decoder(h_z + h_n) if self.use_decoder and self.decoder is not None else None
        return logits, h_z, h_n, x_recon

    def extract_features(self, x):
        h = self.encoder(x)
        if self.use_film and self.film is not None:
            psi = torch.zeros(x.shape[0], 6, device=x.device)
            params = self.film(psi).view(-1, 2, 64, 1)
            gamma, beta = params[:, 0], params[:, 1]
            h = h * gamma + beta
        return self.to_Z(h)

def extract_physics_features(x):
    B, C, T = x.shape
    features = []
    if C >= 3:
        acc = x[:, :3, :]
        gravity_dir = acc.mean(dim=-1)
        gravity_dir = F.normalize(gravity_dir, dim=1, eps=1e-8)
        features.append(gravity_dir)
    if C >= 3:
        acc = x[:, :3, :]
        jerk = torch.diff(acc, dim=-1)
        jerk_mag = jerk.norm(dim=1).mean(dim=-1, keepdim=True)
        features.append(jerk_mag)
    energy = (x ** 2).mean(dim=(1, 2)).unsqueeze(1)
    features.append(energy)
    x_fft = torch.fft.rfft(x, dim=-1)
    power = x_fft.abs() ** 2
    freq_bins = torch.arange(power.shape[-1], device=x.device).float()
    power_mean = power.mean(dim=1)
    freq_center = (power_mean * freq_bins).sum(dim=-1) / (power_mean.sum(dim=-1) + 1e-8)
    freq_center = freq_center.unsqueeze(1)
    features.append(freq_center)
    psi = torch.cat(features, dim=-1)
    return psi

class DoAugmentation:
    @staticmethod
    def apply_random_augmentation(x):
        x = x.clone()
        if torch.rand(1) < 0.5:
            angle = torch.rand(1) * 2 * 3.14159
            rot_matrix = torch.tensor([
                [torch.cos(angle), -torch.sin(angle), 0],
                [torch.sin(angle), torch.cos(angle), 0],
                [0, 0, 1]
            ], device=x.device, dtype=x.dtype)
            if x.shape[1] >= 3:
                acc = x[:, :3, :]
                acc_rot = torch.einsum('ij,bjt->bit', rot_matrix, acc)
                x = torch.cat([acc_rot, x[:, 3:, :]], dim=1)
        if torch.rand(1) < 0.5:
            noise = torch.randn_like(x) * 0.02
            x = x + noise
        if torch.rand(1) < 0.3:
            factor = torch.rand(1) * 0.4 + 0.8
            try:
                B, C, T = x.shape
                new_length = max(1, int(T * factor))
                x_reshaped = x.view(B * C, 1, T)
                x_resized = F.interpolate(x_reshaped, size=new_length, mode='linear', align_corners=False)
                x_resized = F.interpolate(x_resized, size=T, mode='linear', align_corners=False)
                x = x_resized.view(B, C, T)
            except:
                pass
        return x

class CausalHARLoss(nn.Module):
    def __init__(self, use_reconstruction=True, use_invariance=True, use_disentanglement=True):
        super().__init__()
        self.use_reconstruction = use_reconstruction
        self.use_invariance = use_invariance
        self.use_disentanglement = use_disentanglement
        self.ce = nn.CrossEntropyLoss()
        self.mse = nn.MSELoss()
        self.l1 = nn.L1Loss()

    def stft_loss(self, x, x_recon):
        try:
            B, C, T = x.shape
            x_flat = x.view(-1, T)
            x_recon_flat = x_recon.view(-1, T)
            n_fft = min(32, T//2)
            hop_length = n_fft // 2
            win_length = n_fft
            if n_fft < 4:
                return torch.tensor(0.0, device=x.device)
            window = torch.hann_window(win_length, device=x.device)
            x_stft = torch.stft(x_flat, n_fft=n_fft, hop_length=hop_length,
                               win_length=win_length, window=window,
                               return_complex=True, center=False)
            x_recon_stft = torch.stft(x_recon_flat, n_fft=n_fft, hop_length=hop_length,
                                     win_length=win_length, window=window,
                                     return_complex=True, center=False)
            return F.mse_loss(x_stft.abs(), x_recon_stft.abs())
        except:
            return torch.tensor(0.0, device=x.device)

    def hsic_loss(self, z, n, batch_size):
        try:
            z = z.view(batch_size, -1)
            n = n.view(batch_size, -1)
            z_norm = F.normalize(z, dim=1, eps=1e-8)
            n_norm = F.normalize(n, dim=1, eps=1e-8)
            Kz = torch.mm(z_norm, z_norm.t())
            Kn = torch.mm(n_norm, n_norm.t())
            H = torch.eye(batch_size, device=z.device) - 1.0 / batch_size
            hsic_value = torch.trace(torch.mm(torch.mm(Kz, H), torch.mm(Kn, H))) / ((batch_size - 1) ** 2)
            return hsic_value
        except:
            return torch.tensor(0.0, device=z.device)

    def forward(self, logits, y, x, x_recon=None, z=None, n=None, z_prime=None):
        ce_loss = self.ce(logits, y)
        rec_loss = torch.tensor(0.0, device=logits.device)
        if self.use_reconstruction and x_recon is not None:
            rec_loss = 0.5 * self.mse(x_recon, x) + 0.5 * self.stft_loss(x, x_recon)
        inv_loss = torch.tensor(0.0, device=logits.device)
        if self.use_invariance and z is not None and z_prime is not None:
            inv_loss = self.mse(z, z_prime)
        dis_loss = torch.tensor(0.0, device=logits.device)
        if self.use_disentanglement and z is not None and n is not None:
            batch_size = z.shape[0]
            dis_loss = self.hsic_loss(z, n, batch_size)
        total_loss = ce_loss + 0.1 * rec_loss + 0.1 * inv_loss + 0.01 * dis_loss
        return total_loss, ce_loss.item(), rec_loss.item(), inv_loss.item(), dis_loss.item()

def show_confusion_matrix(cm, class_names, title="Confusion Matrix"):
    try:
        cm_norm = cm.astype('float') / (cm.sum(axis=1)[:, np.newaxis] + 1e-8)
        df = pd.DataFrame(cm_norm, index=class_names, columns=class_names)
        annot = pd.DataFrame(cm, index=class_names, columns=class_names)
        plt.figure(figsize=(6, 6))
        sns.heatmap(df, annot=annot.values, fmt="", cmap="Blues", cbar=True,
                    annot_kws={"size": 9}, vmin=0, vmax=1)
        plt.xticks(rotation=90, fontsize=9)
        plt.yticks(rotation=0, fontsize=9)
        for spine in plt.gca().spines.values():
            spine.set_visible(True)
            spine.set_linewidth(0.5)
            spine.set_edgecolor('black')
        plt.xlabel('Predicted Label', fontsize=10)
        plt.ylabel('True Label', fontsize=10)
        plt.title(title, fontsize=12)
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(f"Error plotting confusion matrix: {e}")

def plot_tsne(features, labels, activity_names, dataset_name="Dataset", samples_per_class=400):
    try:
        print(f"Input features shape: {features.shape}, labels shape: {labels.shape}")
        sampled_features, sampled_labels = [], []
        for i in range(len(activity_names)):
            class_mask = (labels == i)
            class_indices = np.where(class_mask)[0]
            if len(class_indices) > 0:
                if len(class_indices) >= samples_per_class:
                    selected_indices = np.random.choice(class_indices, samples_per_class, replace=False)
                else:
                    selected_indices = class_indices
                sampled_features.append(features[selected_indices])
                sampled_labels.append(labels[selected_indices])
        if not sampled_features:
            print("No valid samples found")
            return
        features_sampled = np.vstack(sampled_features)
        labels_sampled = np.concatenate(sampled_labels)
        print(f"Sampled features shape: {features_sampled.shape}, labels shape: {labels_sampled.shape}")

        n_samples_total = features_sampled.shape[0]
        if n_samples_total <= 1:
            print("Not enough samples for t-SNE plot.")
            return

        current_perplexity = min(20, n_samples_total - 1)
        if current_perplexity < 5:
            print(f"Warning: Low sample count ({n_samples_total}) for t-SNE. Perplexity set to {current_perplexity}.")

        features_2d = TSNE(n_components=2, perplexity=current_perplexity, learning_rate=3000, random_state=42).fit_transform(features_sampled)
        colors = [
            "#FF0000", "#FFA500", "#FFFF00", "#008000", "#00FFFF", "#0000FF",
            "#800080", "#FFC0CB", "#A52A2A", "#000000", "#808080", "#FFD700"
        ]
        plt.figure(figsize=(6, 5))
        for i, activity in enumerate(activity_names):
            mask = (labels_sampled == i)
            if np.any(mask):
                plt.scatter(
                    features_2d[mask, 0], features_2d[mask, 1],
                    color=colors[i], marker='o', s=20, alpha=0.4
                )
        handles = [
            plt.Line2D([0], [0], marker='o', color='w', label=activity,
                       markerfacecolor=colors[i], markersize=7)
            for i, activity in enumerate(activity_names)
        ]
        plt.legend(handles=handles, title="Activities", fontsize=9, framealpha=1)
        plt.xlabel("t-SNE Component 1", fontsize=13)
        plt.ylabel("t-SNE Component 2", fontsize=13)
        plt.title(f"t-SNE - {dataset_name}", fontsize=13)
        plt.grid(False)
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(f"Error creating t-SNE plot: {e}")

def compute_model_stats(model, input_shape=(9, 128)):
    try:
        device = next(model.parameters()).device
        dummy_input = torch.randn(1, *input_shape).to(device)
        model.eval()
        with torch.no_grad():
            classifier_original = model.classifier
            model.classifier = nn.Identity()
            macs_backbone, params_backbone = profile(model, inputs=(dummy_input,), verbose=False)
            model.classifier = classifier_original
            features = model.extract_features(dummy_input)
            macs_cls, params_cls = profile(model.classifier, inputs=(features,), verbose=False)
        total_macs_m = (macs_backbone + macs_cls) / 1e6
        total_params_m = (params_backbone + params_cls) / 1e6
        return total_params_m, total_macs_m
    except Exception as e:
        print(f"Error computing model stats: {e}")
        return 0.0, 0.0

def measure_inference_time(model, input_size=(1, 9, 128), device='cuda', n_runs=100, warmup=10):
    try:
        model.eval()
        inputs = torch.randn(input_size).to(device)
        psi = torch.zeros(input_size[0], 6, device=device)
        with torch.inference_mode():
            for _ in range(warmup):
                _ = model(inputs, psi)
        if device == 'cuda' and torch.cuda.is_available():
            torch.cuda.synchronize()
        start = time.time()
        with torch.inference_mode():
            for _ in range(n_runs):
                _ = model(inputs, psi)
        if device == 'cuda' and torch.cuda.is_available():
            torch.cuda.synchronize()
        end = time.time()
        return (end - start) / n_runs * 1000
    except Exception as e:
        print(f"Error measuring inference time: {e}")
        return 0.0

class Trainer:
    def __init__(self, model, loss_fn, optimizer, scheduler, device, activity_names, save_dir):
        self.model = model.to(device)
        self.loss_fn = loss_fn
        self.optimizer = optimizer
        self.scheduler = scheduler
        self.device = device
        self.activity_names = activity_names
        self.save_dir = save_dir
        os.makedirs(save_dir, exist_ok=True)
        self.best_f1 = 0
        self.best_metrics = {}

    def train_epoch(self, train_loader, val_loader, epoch):
        self.model.train()
        train_loss = 0
        train_correct = 0
        train_total = 0
        ce_losses = []
        rec_losses = []
        inv_losses = []
        dis_losses = []
        for batch_idx, (x, y) in enumerate(train_loader):
            try:
                x, y = x.to(self.device), y.to(self.device)
                psi = extract_physics_features(x)
                x_aug = DoAugmentation.apply_random_augmentation(x)
                psi_aug = extract_physics_features(x_aug)
                logits, z, n, x_recon = self.model(x, psi)
                _, z_prime, _, _ = self.model(x_aug, psi_aug)
                loss, ce, rec, inv, dis = self.loss_fn(logits, y, x, x_recon, z, n, z_prime)
                self.optimizer.zero_grad()
                loss.backward()
                torch.nn.utils.clip_grad_norm_(self.model.parameters(), 1.0)
                self.optimizer.step()
                train_loss += loss.item()
                pred = logits.argmax(dim=1)
                train_correct += (pred == y).sum().item()
                train_total += y.size(0)
                ce_losses.append(ce)
                rec_losses.append(rec)
                inv_losses.append(inv)
                dis_losses.append(dis)
            except Exception as e:
                print(f"Error in training step {batch_idx}: {e}")
                continue
        train_acc = train_correct / max(train_total, 1)
        avg_train_loss = train_loss / max(len(train_loader), 1)
        avg_ce = np.mean(ce_losses) if ce_losses else 0
        avg_rec = np.mean(rec_losses) if rec_losses else 0
        avg_inv = np.mean(inv_losses) if inv_losses else 0
        avg_dis = np.mean(dis_losses) if dis_losses else 0
        self.model.eval()
        val_loss = 0
        val_correct = 0
        val_total = 0
        all_preds, all_labels, all_z = [], [], []
        with torch.no_grad():
            for batch_idx, (x, y) in enumerate(val_loader):
                try:
                    x, y = x.to(self.device), y.to(self.device)
                    psi = extract_physics_features(x)
                    logits, z, n, x_recon = self.model(x, psi)
                    loss, _, _, _, _ = self.loss_fn(logits, y, x, x_recon, z, n)
                    val_loss += loss.item()
                    pred = logits.argmax(dim=1)
                    val_correct += (pred == y).sum().item()
                    val_total += y.size(0)
                    all_preds.extend(pred.cpu().numpy())
                    all_labels.extend(y.cpu().numpy())
                    z_flat = z.view(z.size(0), -1).cpu().numpy()
                    all_z.extend(z_flat)
                except Exception as e:
                    print(f"Error in validation step {batch_idx}: {e}")
                    continue
        val_acc = val_correct / max(val_total, 1)
        avg_val_loss = val_loss / max(len(val_loader), 1)
        if epoch % 25 == 0:
            print(f"Epoch {epoch} | "
                  f"Train Loss: {avg_train_loss:.6f} | Train Acc: {train_acc:.4f} | "
                  f"Val Loss: {avg_val_loss:.6f} | Val Acc: {val_acc:.4f}")
            print(f" CE: {avg_ce:.6f} | Rec: {avg_rec:.6f} | Inv: {avg_inv:.6f} | Dis: {avg_dis:.6f}")
        if all_preds and all_labels:
            f1_macro = f1_score(all_labels, all_preds, average='macro', zero_division=0)
            if f1_macro > self.best_f1:
                self.best_f1 = f1_macro
                self.best_metrics = {
                    'acc': val_acc,
                    'f1_macro': f1_macro,
                    'f1_weighted': f1_score(all_labels, all_preds, average='weighted', zero_division=0),
                    'precision': precision_score(all_labels, all_preds, average='macro', zero_division=0),
                    'recall': recall_score(all_labels, all_preds, average='macro', zero_division=0),
                    'params': compute_model_stats(self.model)[0],
                    'flops': compute_model_stats(self.model)[1],
                    'inf_ms': measure_inference_time(self.model, device=self.device),
                    'cm': confusion_matrix(all_labels, all_preds),
                    'z': np.array(all_z),
                    'preds': np.array(all_preds),
                    'labels': np.array(all_labels)
                }
        if self.scheduler:
            self.scheduler.step()

    def get_best_results(self):
        return self.best_metrics

    def show_final_results(self, config_name):
        if not self.best_metrics:
            print("No best metrics available")
            return
        print(f"\n{'='*50}")
        print(f"FINAL RESULTS FOR {config_name}")
        print(f"{'='*50}")
        print(f"Accuracy: {self.best_metrics['acc']:.4f}")
        print(f"F1 (macro): {self.best_metrics['f1_macro']:.4f}")
        print(f"F1 (weighted): {self.best_metrics['f1_weighted']:.4f}")
        print(f"Precision: {self.best_metrics['precision']:.4f}")
        print(f"Recall: {self.best_metrics['recall']:.4f}")
        print(f"Parameters (M): {self.best_metrics['params']:.2f}")
        print(f"FLOPs (M): {self.best_metrics['flops']:.2f}")
        print(f"Inference (ms): {self.best_metrics['inf_ms']:.2f}")
        print(f"\nConfusion Matrix for {config_name}:")
        show_confusion_matrix(self.best_metrics['cm'], self.activity_names,
                             f"Confusion Matrix - {config_name}")
        print(f"\nt-SNE Visualization for {config_name}:")
        plot_tsne(self.best_metrics['z'], self.best_metrics['labels'],
                 self.activity_names, config_name, samples_per_class=400)

class RobustnessTests:
    @staticmethod
    def apply_rotation(x_batch, degrees):
        try:
            device = x_batch.device
            dtype = x_batch.dtype
            rad = torch.tensor(degrees * np.pi / 180.0, device=device, dtype=dtype)

            axis = F.normalize(torch.rand(3, device=device, dtype=dtype), dim=0)
            K = torch.zeros((3, 3), device=device, dtype=dtype)
            K[0, 1] = -axis[2]
            K[0, 2] = axis[1]
            K[1, 0] = axis[2]
            K[1, 2] = -axis[0]
            K[2, 0] = -axis[1]
            K[2, 1] = axis[0]

            cos_rad = torch.cos(rad)
            sin_rad = torch.sin(rad)
            I = torch.eye(3, device=device, dtype=dtype)
            R = I + sin_rad * K + (1 - cos_rad) * torch.mm(K, K)

            x_aug_list = []
            if x_batch.shape[1] >= 3:
                x_aug_list.append(torch.einsum('ij,bjt->bit', R, x_batch[:, 0:3, :]))
            if x_batch.shape[1] >= 6:
                x_aug_list.append(torch.einsum('ij,bjt->bit', R, x_batch[:, 3:6, :]))
            if x_batch.shape[1] >= 9:
                x_aug_list.append(torch.einsum('ij,bjt->bit', R, x_batch[:, 6:9, :]))

            if len(x_aug_list) == 3:
                return torch.cat(x_aug_list, dim=1)
            else:
                return x_batch
        except Exception as e:
            return x_batch

    @staticmethod
    def apply_gaussian_noise(x_batch, percentage):
        try:
            noise_std = (percentage / 100.0) * torch.std(x_batch)
            noise = torch.randn_like(x_batch) * noise_std
            return x_batch + noise
        except Exception as e:
            return x_batch

    @staticmethod
    def apply_elimination(x_batch, percentage):
        try:
            B, C, T = x_batch.shape
            num_to_zero = int(T * (percentage / 100.0))
            if num_to_zero == 0: return x_batch

            indices = torch.stack([torch.randperm(T, device=x_batch.device)[:num_to_zero] for _ in range(B)])
            x_aug = x_batch.clone()

            batch_indices = torch.arange(B, device=x_batch.device).view(-1, 1).expand_as(indices)

            x_aug[batch_indices, :, indices] = 0.0
            return x_aug
        except Exception as e:
            return x_batch

    @staticmethod
    def apply_scaling(x_batch, percentage):
        try:
            factor = 1.0 + (percentage / 100.0)
            return x_batch * factor
        except Exception as e:
            return x_batch

    @staticmethod
    def apply_time_warping(x_batch, percentage):
        try:
            B, C, T = x_batch.shape
            factor = 1.0 + (torch.rand(1).item() * (percentage / 100.0) * 2) - (percentage / 100.0)
            new_length = max(1, int(T * factor))
            if new_length == T: return x_batch

            x_reshaped = x_batch.view(B * C, 1, T)
            x_resized = F.interpolate(x_reshaped, size=new_length, mode='linear', align_corners=False)
            x_resized = F.interpolate(x_resized, size=T, mode='linear', align_corners=False)
            return x_resized.view(B, C, T)
        except Exception as e:
            return x_batch

    @staticmethod
    def apply_freq_masking(x_batch, percentage):
        try:
            B, C, T = x_batch.shape
            x_fft = torch.fft.rfft(x_batch, dim=-1)
            n_freqs = x_fft.shape[-1]
            mask_width = int(n_freqs * (percentage / 100.0))
            if mask_width == 0: return x_batch

            mask_start = torch.randint(0, n_freqs - mask_width + 1, (B, C), device=x_batch.device)
            mask_indices = torch.arange(n_freqs, device=x_batch.device).view(1, 1, -1)

            mask = (mask_indices >= mask_start.unsqueeze(-1)) & (mask_indices < (mask_start + mask_width).unsqueeze(-1))
            x_fft_masked = x_fft.masked_fill(mask, 0)

            return torch.fft.irfft(x_fft_masked, n=T, dim=-1)
        except Exception as e:
            return x_batch

    @staticmethod
    def apply_freq_filtering(x_batch, percentage):
        try:
            B, C, T = x_batch.shape
            x_fft = torch.fft.rfft(x_batch, dim=-1)
            n_freqs = x_fft.shape[-1]
            keep_width = int(n_freqs * (percentage / 100.0))
            if keep_width == 0: return torch.zeros_like(x_batch)
            if keep_width >= n_freqs: return x_batch

            keep_start = torch.randint(0, n_freqs - keep_width + 1, (B, C), device=x_batch.device)
            mask_indices = torch.arange(n_freqs, device=x_batch.device).view(1, 1, -1)

            mask = (mask_indices < keep_start.unsqueeze(-1)) | (mask_indices >= (keep_start + keep_width).unsqueeze(-1))
            x_fft_filtered = x_fft.masked_fill(mask, 0)

            return torch.fft.irfft(x_fft_filtered, n=T, dim=-1)
        except Exception as e:
            return x_batch

def run_robustness_evaluation(trained_models_map, device, activity_names):
    print(f"\n{'='*80}")
    print("RUNNING ROBUSTNESS EVALUATION")
    print(f"{'='*80}")

    robustness_results = []

    test_scenarios = [
        ("Rotation", [5, 10, 15, 20, 25], RobustnessTests.apply_rotation),
        ("Gaussian Noise", [5, 10, 15, 20, 25, 30], RobustnessTests.apply_gaussian_noise),
        ("Elimination", [5, 10, 15, 20, 25, 30], RobustnessTests.apply_elimination),
        ("Scaling", [5, 10, 15, 20, 25, 30], RobustnessTests.apply_scaling),
        ("Time Warping", [5, 10, 15, 20, 25, 30], RobustnessTests.apply_time_warping),
        ("Freq Masking", [5, 10, 15, 20, 25, 30], RobustnessTests.apply_freq_masking),
        ("Freq Filtering", [5, 10, 15, 20, 25, 30], RobustnessTests.apply_freq_filtering)
    ]

    for name, (model, test_dataset) in trained_models_map.items():
        print(f"\n--- Testing Model: {name} ---")
        test_loader = DataLoader(test_dataset, batch_size=512, shuffle=False, num_workers=2, pin_memory=True)
        model.eval()

        for test_name, intensities, aug_fn in test_scenarios:
            print(f"  Test: {test_name}")
            for intensity in intensities:
                all_preds = []
                all_labels = []
                with torch.no_grad():
                    for x_batch, y_batch in test_loader:
                        x_batch, y_batch = x_batch.to(device), y_batch.to(device)

                        x_aug = aug_fn(x_batch.clone(), intensity)

                        psi_aug = extract_physics_features(x_aug)

                        logits, _, _, _ = model(x_aug, psi_aug)
                        preds = logits.argmax(dim=1)

                        all_preds.append(preds.cpu())
                        all_labels.append(y_batch.cpu())

                all_preds_np = torch.cat(all_preds).numpy()
                all_labels_np = torch.cat(all_labels).numpy()

                acc = accuracy_score(all_labels_np, all_preds_np)
                f1 = f1_score(all_labels_np, all_preds_np, average='macro', zero_division=0)

                robustness_results.append({
                    "Model": name,
                    "Test": test_name,
                    "Intensity": f"{intensity}%" if test_name != "Rotation" else f"{intensity}",
                    "Acc": round(acc, 4),
                    "F1": round(f1, 4)
                })

    print(f"\n{'='*80}")
    print("FINAL ROBUSTNESS TEST RESULTS")
    print(f"{'='*80}")
    test_df = pd.DataFrame(robustness_results)
    display(test_df)

    print("\n--- Pivot Table (Accuracy) ---")
    display(test_df.pivot_table(index=["Test", "Intensity"], columns="Model", values="Acc"))
    print (test_df.pivot_table(index=["Test", "Intensity"], columns="Model", values="Acc"))
    print("\n--- Pivot Table (F1-Score) ---")
    display(test_df.pivot_table(index=["Test", "Intensity"], columns="Model", values="F1"))
    print (test_df.pivot_table(index=["Test", "Intensity"], columns="Model", values="F1"))

def run_ablation_study(data_root=None):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")

    if data_root is None:
        data_root = "./UCI HAR Dataset"
        if not os.path.exists(data_root):
            print(f"Data directory {data_root} not found. Please provide correct path.")
            try:
                from google.colab import drive
                drive.mount('/content/drive')
                data_root = "/content/drive/MyDrive/HAR_Dataset/UCI"
                if not os.path.exists(data_root):
                     print(f"Colab data directory {data_root} not found either. Exiting.")
                     return
            except ImportError:
                 print("Not in Colab, and local path not found. Exiting.")
                 return

    try:
        X_train_raw, y_train_raw, _, _ = load_split(data_root, "train")
        X_test_raw, y_test_raw, _, activity_names = load_split(data_root, "test")

        indices_train_val = np.arange(len(y_train_raw))
        np.random.shuffle(indices_train_val)
        val_size = int(0.2 * len(y_train_raw))
        val_idx, train_orig_idx = indices_train_val[:val_size], indices_train_val[val_size:]

        X_train_orig = X_train_raw[train_orig_idx]
        y_train_orig = y_train_raw[train_orig_idx]

        X_val_split = X_train_raw[val_idx]
        y_val_split = y_train_raw[val_idx]

        configs = [
            ("Full", True, True, True, True),
            ("w_o_FiLM", False, True, True, True),
            ("w_o_Inv", True, True, False, True),
            ("w_o_Dis", True, True, True, False),
            ("w_o_Dec", True, False, True, True),
            ("Baseline", False, False, False, False),
        ]
        results = []
        trained_models_map = {}

        for name, use_film, use_decoder, use_inv, use_dis in configs:
            print(f"\n{'='*60}")
            print(f"RUNNING: {name}")
            print(f"{'='*60}")
            save_dir = f"./results/{name}"
            os.makedirs(save_dir, exist_ok=True)

            train_dataset = HARDataset(X_train_orig, y_train_orig, normalize=True)
            mean, std = train_dataset.mean, train_dataset.std

            val_dataset = HARDataset(X_val_split, y_val_split, normalize=False)
            val_dataset.set_stats(mean, std)

            test_dataset = HARDataset(X_test_raw, y_test_raw, normalize=False)
            test_dataset.set_stats(mean, std)

            train_loader = DataLoader(train_dataset, batch_size=512, shuffle=True, num_workers=2, pin_memory=True, drop_last=True)
            val_loader = DataLoader(val_dataset, batch_size=512, shuffle=False, num_workers=2, pin_memory=True)
            test_loader = DataLoader(test_dataset, batch_size=512, shuffle=False, num_workers=2, pin_memory=True)

            try:
                model = CausalManifoldHAR(use_film=use_film, use_decoder=use_decoder)
                loss_fn = CausalHARLoss(use_reconstruction=use_decoder, use_invariance=use_inv, use_disentanglement=use_dis)
                optimizer = torch.optim.AdamW(model.parameters(), lr=1e-3, weight_decay=1e-5)
                scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=200)
                trainer = Trainer(model, loss_fn, optimizer, scheduler, device, activity_names, save_dir)

                for epoch in range(1, 201):
                    trainer.train_epoch(train_loader, val_loader, epoch)

                model.eval()
                all_preds, all_labels, all_z = [], [], []
                test_loss = 0
                test_correct = 0
                test_total = 0

                with torch.no_grad():
                    for x, y in test_loader:
                        x, y = x.to(device), y.to(device)
                        psi = extract_physics_features(x)
                        logits, z, n, x_recon = model(x, psi)
                        loss, _, _, _, _ = loss_fn(logits, y, x, x_recon, z, n)
                        test_loss += loss.item()
                        pred = logits.argmax(dim=1)
                        test_correct += (pred == y).sum().item()
                        test_total += y.size(0)
                        all_preds.extend(pred.cpu().numpy())
                        all_labels.extend(y.cpu().numpy())
                        z_flat = z.view(z.size(0), -1).cpu().numpy()
                        all_z.extend(z_flat)

                if test_total == 0:
                    print(f"Config {name} FAILED: No samples in test_loader.")
                    continue

                test_acc = test_correct / test_total
                f1_macro = f1_score(all_labels, all_preds, average='macro', zero_division=0)
                f1_weighted = f1_score(all_labels, all_preds, average='weighted', zero_division=0)
                precision = precision_score(all_labels, all_preds, average='macro', zero_division=0)
                recall = recall_score(all_labels, all_preds, average='macro', zero_division=0)
                cm = confusion_matrix(all_labels, all_preds)

                best_metrics = {
                    'acc': test_acc,
                    'f1_macro': f1_macro,
                    'f1_weighted': f1_weighted,
                    'precision': precision,
                    'recall': recall,
                    'params': compute_model_stats(model)[0],
                    'flops': compute_model_stats(model)[1],
                    'inf_ms': measure_inference_time(model, device=device),
                    'cm': cm,
                    'z': np.array(all_z),
                    'preds': np.array(all_preds),
                    'labels': np.array(all_labels)
                }

                results.append({
                    'Config': name,
                    'F1(m)': round(f1_macro, 4),
                    'Acc': round(test_acc, 4),
                    'Prec': round(precision, 4),
                    'Rec': round(recall, 4),
                    'Params(M)': round(best_metrics['params'], 2),
                    'FLOPs(M)': round(best_metrics['flops'], 2),
                    'Inf(ms)': round(best_metrics['inf_ms'], 2),
                })

                trainer.best_metrics = best_metrics
                trainer.best_f1 = f1_macro
                trainer.show_final_results(name)

                if name in ["Full", "Baseline"]:
                    trained_models_map[name] = (model, test_dataset)

            except Exception as e:
                print(f"Error in config {name}: {e}")
                continue

        if results:
            print(f"\n{'='*80}")
            print(f"{'='*80}")
            rst_df = pd.DataFrame(results)
            display(rst_df)
        else:
            print("No results were generated. Check for errors above.")

        if "Full" in trained_models_map and "Baseline" in trained_models_map:
            run_robustness_evaluation(trained_models_map, device, activity_names)
        else:
            print("Could not run robustness test: 'Full' or 'Baseline' model not trained.")

    except Exception as e:
        print(f"Error in ablation study: {e}")

if __name__ == "__main__":
    run_ablation_study(data_root=None)
```
